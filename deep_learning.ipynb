## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

- EIN and NAME—Identification columns
- APPLICATION_TYPE—Alphabet Soup application type
- AFFILIATION—Affiliated sector of industry
- CLASSIFICATION—Government organization classification
- USE_CASE—Use case for funding
- ORGANIZATION—Organization type
- STATUS—Active status
- INCOME_AMT—Income classification
- SPECIAL_CONSIDERATIONS—Special considerations for application
- ASK_AMT—Funding amount requested
- IS_SUCCESSFUL—Was the money used effectively

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(columns = ['EIN', 'NAME'])
```

```python
# Determine the number of unique values in each column.
application_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
AT_vcount = application_df['APPLICATION_TYPE'].value_counts()
AT_vcount[AT_vcount > 200]
```

```python
# Choose a cutoff value and create a list of application types to be replaced
cutoff = AT_vcount[AT_vcount > 200]
other = AT_vcount[AT_vcount < 200]

other.index

# Replace in dataframe
for app in other.index:
  application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
class_vcounts = application_df['CLASSIFICATION'].value_counts()
class_vcounts
```

```python
# CLASSIFICATION value counts >1
class_vcounts[class_vcounts > 1]
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
class_other = class_vcounts[class_vcounts < 1500]

# Replace in dataframe
for cls in class_other.index:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
application_df.dtypes
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
dummies_df = pd.get_dummies(application_df.drop(columns=['STATUS', 'ASK_AMT', 'IS_SUCCESSFUL']), dtype = int)
final_data = dummies_df.join(application_df[['STATUS', 'ASK_AMT', 'IS_SUCCESSFUL']])
```

```python
final_data.head()
```

```python
# Split our preprocessed data into our features and target arrays
features = final_data.copy().drop(columns = 'IS_SUCCESSFUL')
target = final_data['IS_SUCCESSFUL']

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size = 0.3, random_state = 42)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Finding the best model

```python
# Create a method that creates a new Sequential model with hyperparameter options
def create_model(hp):
    nn_model = tf.keras.models.Sequential()

    # Allow kerastuner to decide which activation function to use in hidden layers
    activation = hp.Choice('activation',['relu', 'tanh'])

    # Allow kerastuner to decide number of neurons in first layer
    nn_model.add(tf.keras.layers.Dense(units=hp.Int('first_units',
        min_value=1,
        max_value=30,
        step=5), activation=activation, input_dim=len(features.columns)))

    # Allow kerastuner to decide number of hidden layers and neurons in hidden layers
    for i in range(hp.Int('num_layers', 1, 5)):
        nn_model.add(tf.keras.layers.Dense(units=hp.Int('units_' + str(i),
            min_value=1,
            max_value=30,
            step=5),
            activation=activation))

    nn_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

    # Compile the model
    nn_model.compile(loss="binary_crossentropy", optimizer='adam', metrics=["accuracy"])

    return nn_model
```

## Compile, Train and Evaluate the Model

```python
len(features.columns)
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
model = tf.keras.models.Sequential()

# First hidden layer
model.add(tf.keras.layers.Dense(units = 26, activation = 'tanh', input_dim = len(features.columns)))

# Second hidden layer
model.add(tf.keras.layers.Dense(units = 21, activation = 'tanh'))

# Third hidden layer
# model.add(tf.keras.layers.Dense(units = 21, activation = 'tanh'))

# Output layer
model.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid'))

# Check the structure of the model
model.summary()
```

```python
# Compile the model
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
# Train the model
model.fit(X_train_scaled, y_train, epochs = 100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
model.save('tf_model.h5')
```
